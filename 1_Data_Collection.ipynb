# 1. **Data Collection**

```python
import json
import requests
from getpass import getpass
from bs4 import BeautifulSoup
```

## Naver Open API



Naver Open API: https://developers.naver.com/products/service-api/datalab/datalab.md

Naver Open API for news: https://developers.naver.com/docs/serviceapi/search/news/news.md

```python
client_id = getpass('Naver Open API client ID: ')
client_secret = getpass('Naver Open API client secret: ')
```

```python
headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret
}

url = 'https://openapi.naver.com/v1/search/news?query=소비자물가'
raw = requests.get(url, headers=headers)
json.loads(raw.text)['items'][:3]
```

## Crawling news articles

```python
# Naver news article URL('https://n.news.naver.com/mnews/article/...')
url = 'https://n.news.naver.com/mnews/article/001/0014265761?sid=104'
```

```python
# set header for crawling
header = {'User-Agent': 'Mozilla/5.0'}
raw = requests.get(url, headers=header)
print(raw.text[0:1000])
```

```python
html = BeautifulSoup(raw.text, 'html.parser')
# get the title
title = html.find('h2', class_='media_end_head_headline')
if title is None:
    print('No title found')
else:
    title = title.get_text().replace('\n', '')
    print(title)
```

```python
# get the content
content = html.find('div', id='newsct_article').get_text()
if content is None:
    print('No content found')
else:
    print(content)
```
